<h1 align="center"> Basic Plotting with Matplotlib </h1>

`Matplotlib` is a plotting library for Python. It provides a very quick way to visualize data from Python. Matplotlib was originally written by John D. Hunter. Started as a small project, today Matplotlib is used for creating publication quality plots and figures.

Matplotlib is heavily reliant on packages such as Numpy (we'll talk about that in Week 5 and 6 of our course) and its then that you will truly understand the beauty of matplotlib. For now, lets start making some basic plots.

```python
# To make sure the figures are plotted in the web page and not as a separate pop-up, we invoke jupyter notebooks magic function
%matplotlib inline
# Import matplotlib's pyplot module
import matplotlib.pyplot as plt
```

Lets create two lists, one containing a number and the other containing square of that number

```python
numbers = range(1, 10)
print numbers
```

```python
squares = map(lambda x: x**2, numbers)
print squares
```

It is important to make sure that the two `list`s (or arrays as we'll see in future) you use for plotting have the same dimensions, or `matplotlib` will raise an exception when you try to plot them.

Now lets try to plot number vs squares:

```python
fig = plt.figure()
ax = fig.add_subplot(111) # nrows, mcols, plot_number
ax.plot(numbers, squares)
```

All plotting is done through the Figure object. They are the "canvas" of your plot.
subplot(nrows, mcols, plot_number) is a function that adds a subplot on your figure at n<sup>th</sup> row, m<sup>th</sup> column and at `plot_number`. Don't worry if you don't understand everything right away. Things will become more clear once you reach Week 5 and 6

Great! Your first plot. However if some one looks at the figure, they wouldn't understand what the plot is about. Lets try to make sense for them. *(Remember, as a scientist you need to be able to tell a story with your plot)*

**Adding Axes**:

A figure without axis labels is useless to anyone other than the person who created it.  Let’s add some to our plot.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
```

**Adding title**:

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
```

```python
# Lets add another set of data points.. Cube of the numbers
cubes = map(lambda x: x**3, numbers)
print cubes
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
ax.plot(numbers, cubes)
```

**Adding Legends and Formatting**:



Let’s add a legend in the upper left corner. This only requires adding the keyword argument label (that will be used in the legend box) to the plot commands.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares, label="Squares")
ax.plot(numbers, cubes, label="Cubes")
ax.legend(loc="upper left")
```

Lets also use different colors and linestyles

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares, label="Squares", marker='o', linestyle='--', color='red')
ax.plot(numbers, cubes, label="Cubes", marker='+', linestyle=':', color='blue')
ax.legend(loc="upper left")
```

Matplotlib's `plot` does line plotting by default. Lets try a bar plot

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red')
ax.bar(numbers, cubes, label="Cubes", color='blue', alpha=0.2) #alpha makes plots translucent. Its value ranges from 0 - 1
ax.legend(loc="upper left")
```

How about combining line and bar plots?

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red')
ax.plot(numbers, squares)
ax.bar(numbers, cubes, label="Cubes", color='blue', alpha=0.2) #alpha makes plots translucent. Its value ranges from 0 - 1
ax.plot(numbers, cubes)
ax.legend(loc="upper left")
```

See how the lines are touching the edge of bar plot. It would look good if its in the center.

We can do that by adding a keyword argument `align` in bar method:

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red', align='center')
ax.plot(numbers, squares)
ax.bar(numbers, cubes, label="Cubes", color='blue', align='center', alpha=0.2) #alpha makes plots translucent. Its value ranges from 0 - 1
ax.plot(numbers, cubes)
ax.legend(loc="upper left")
```
